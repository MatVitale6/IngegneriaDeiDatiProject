```python
import os
import json
import re

def generate_id(paper_id, table_id, claim_id, spec_name):
    return f"{paper_id}_{table_id}_{claim_id}_{spec_name.lower().replace(' ', '_')}"

def parse_claim_file(file_path):
    claims = {}
    
    with open(file_path, 'r') as file:
        for line in file:
            if line.strip():
                print(f"Processing line: {line.strip()}")

                claim_match = re.match(r'Claim (\d+): \|{([^}]+)\}, ([^,]+), ([\d\.Â±]+)\|', line.strip())
                
                if claim_match:
                    claim_id = int(claim_match.group(1))  
                    claim_content = claim_match.group(2)  

                    print(f"Found claim {claim_id} with content: {claim_content}")

                    key_value_pairs = re.findall(r'([^,]+),\s*([^|]+)', claim_content)
                    claim_data = {}
                    
                    for key, value in key_value_pairs:
                        claim_data[key.strip()] = value.strip()

                    if claim_data:
                        claims[claim_id] = claim_data

    return claims

claims_directory = 'claims\\claims_GPT-4o'
all_claims = {}


for file_name in os.listdir(claims_directory):
    if file_name.endswith('.txt'):  
        file_path = os.path.join(claims_directory, file_name)
        
        claims = parse_claim_file(file_path)
        
        if claims:
            all_claims[file_name] = claims
            output_file_path = os.path.join('output_claims', f'{file_name}_claims.json')
            
            os.makedirs(os.path.dirname(output_file_path), exist_ok=True)
            
            with open(output_file_path, 'w') as json_file:
                json.dump(claims, json_file, indent=4)

            print(f"Claims for {file_name} saved in {output_file_path}")

print("Processing completed.")
```

```python
def gather_field_names(claims):
    field_names = set()
    for claim in claims.values():
        print(f"Processing claim: {claim}")  # Stampa ogni claim
        for key in claim.keys():
            print(f"Checking key: {key}")  # Stampa ogni chiave
            if isinstance(key, str):  # Assicurati che sia una stringa
                field_names.add(key.strip())
    return field_names

field_names = gather_field_names(all_claims)
print(field_names)
```
